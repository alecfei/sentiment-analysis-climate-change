```python
# import libraries needed
import datetime 
import pandas as pd
import snscrape.modules.twitter as snstwitter
```

```python
query = "(#weather OR 'weather') lang:en since:2022-05-01 until:2023-05-01"
tweets = []

for i, tweet in enumerate(snstwitter.TwitterSearchScraper(query).get_items()):
    if i > 900000:
        break
    else:
        tweets.append([tweet.date, tweet.user.username, tweet.user.location, 
                      tweet.rawContent, tweet.likeCount, tweet.retweetCount, 
                      tweet.user.followersCount, tweet.replyCount]) 
weather_tweets = pd.DataFrame(tweets, columns = ["date", "username", "user_location",
                                                 "tweet", "num_of_like", "num_of_retweet",
                                                 "num_of_followers", "num_of_reply"])

weather_tweets.to_csv("Weather tweets.csv")
```

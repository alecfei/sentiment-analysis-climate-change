## Visualisation of the benchmark testing results of the four databases

<span style="font-size: 12pt;">This notebook diplays the comparison among the four databases chosen based on the benchmarking results by YCSB.</span>

<br>

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Workloada

```python
workloada = pd.DataFrame({
    'Database': ['Cassandra', 'MongoDB', 'HBase', 'MySQL'],
    'Throughput(ops/sec)': [288.08, 765.99, 329.44, 56.26],
    'AverageLatency(us)_READ': [2931.99, 1073.73, 2188.16, 1636.60],
    'AverageLatency(us)_INSERT': [2922.64, 889.24, 2388.66, 27217.87],
    'AverageLatency(us)_UPDATE': [3007.70, 1143.55, 4248.21, 29747.15]
})

workloada
```

```python
def plot_barchart(workload, suptitle, y_column, ylabel, ylim):
    # Set the figure size and create subplots
    fig, axes = plt.subplots(figsize = (6, 6))

    # Set the title of the figure
    fig.suptitle(suptitle)

    # Set the style of the seaborn plot
    sns.set_theme(style = "whitegrid")

    # Sort the ranks based on the specified y-axis column
    ranks = workload[y_column].sort_values().index

    # Define a color palette for the bars
    palette = sns.color_palette("Paired", len(ranks))

    # Create the bar plot
    sns.barplot(ax = axes, data = workload, x = "Database", y = y_column,
                linewidth = 3, hue = "Database", 
                palette = np.array(palette)[ranks], dodge=False)

    # Set the labels for x and y axes
    axes.set_xlabel("Database")
    axes.set_ylabel(ylabel)

    # Add labels to the bars
    for container in axes.containers:
        axes.bar_label(container)

    # Add a legend to the plot
    plt.legend(title = "Database", bbox_to_anchor = (1, 1), 
               frameon =True, ncol = 2, 
               markerfirst = False, shadow = True)

    # Set the y-axis limits
    axes.set_ylim(ylim)

    # Display the plot
    plt.show()
```

```python
plot_barchart(workloada, suptitle = "Comparison of Throughput in workloada", 
              y_column = "Throughput(ops/sec)", ylabel = "Throughput(ops/sec)", 
              ylim=(0, max(workloada["Throughput(ops/sec)"])))
```

```python
plot_barchart(workloada, suptitle = "Comparison of Reading AverageLatency in workloada", 
              y_column = "AverageLatency(us)_READ", ylabel = "AverageLatency(us)", 
              ylim = (0, max(workloada["AverageLatency(us)_READ"])))
```

```python
plot_barchart(workloada, suptitle = "Comparison of Inserting AverageLatency in workloada", 
              y_column = "AverageLatency(us)_INSERT", ylabel = "AverageLatency(us)", 
              ylim = (0, max(workloada["AverageLatency(us)_INSERT"])))
```

```python
plot_barchart(workloada, suptitle = "Comparison of Updating AverageLatency in workloada", 
              y_column = "AverageLatency(us)_UPDATE", ylabel = "AverageLatency(us)", 
              ylim = (0, max(workloada["AverageLatency(us)_UPDATE"])))
```
